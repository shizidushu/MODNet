```python
import os
import glob
import cv2
import numpy as np
from PIL import Image
```

```python
mask_path_pattern = os.path.sep.join(["/home/ubuntu/data/workspace/deeplabv3_plus/people_segmentation", "masks", "*"])
```

```python
mask_paths = sorted(glob.glob(mask_path_pattern))
```

```python
samples = []
for mask_path in mask_paths:
    img_path = os.path.splitext(mask_path.replace('masks', 'images'))[0] + '.jpg'
    if not os.path.exists(img_path):
        img_path = img_path.replace('.jpg', '.png')
    samples.append({
        "img_path": img_path,
        "mask_path": mask_path
    })
```
